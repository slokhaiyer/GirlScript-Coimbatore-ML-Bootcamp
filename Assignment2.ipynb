```
#tax payment
import random
sales = [random.uniform(100,1000000) for i in range(12)]
profit =[random.uniform(100,100000) for i in range(12)]
print("Sales List: ",sales)
print("Profit List: ",profit)

#calculate profit ratio
#formula = (profit/sales)*100
profit_ratio =  [(i / j )*100 for i, j in zip(profit, sales)]
print("Profit Ratio: ",profit_ratio)

#calculate mean sales and mean profit
mean_sales = sum(sales)/12
print("Mean sales is " ,mean_sales)
mean_profit = sum(profit)/12

#good months : months above mean
print("Good months (months with above average profit) Note: The numbers represent the months number eg: 0 implies January and so on: ")
for g in profit:
  if(g > mean_profit):
    print(profit.index(g))

#bad months : months below mean
print("Bad months (months with below average profit) Note: The numbers represent the months number eg: 0 implies January and so on: ")
for b in profit:
  if(b < mean_profit):
    print(profit.index(b))
#best month : max profit
Max_profit = max(profit)
print("Best month, ie month with maximum profit is: ",profit.index(Max_profit), "with a profit of",Max_profit)
#worst month : min profit
Min_profit = min(profit)
print("Worst month, ie month with minimum profit is: ",profit.index(Min_profit), "with a profit of",Min_profit)
#worst month : min profit after tax
#(sales - profit)/sales
profit_after_tax = [(a - b)/a for a, b in zip(sales, profit)]
worstPAT = min(profit_after_tax)
print("Worst month, ie month with worst profit after tax is: ",profit_after_tax.index(worstPAT), "with a profit of",worstPAT)



```
