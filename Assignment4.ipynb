```
#Write a Python program to plot two or more lines with different styles.
import matplotlib.pyplot as plt

x1 = [1,3,5]
y1 = [2,4,6]
x2 = [6,4,2]
y2 = [5,3,1]
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.plot(x1,y1, color='b', linewidth = 3,  label = 'line1', linestyle='dotted')
plt.plot(x2,y2, color='g', linewidth = 3,  label = 'line2', linestyle='dashed')
plt.title("Plot with two or more lines with different styles")
plt.legend()
plt.show()
```

```
#Write a Python program to plot two or more lines and set the line markers.
x = [1,3,5]
y = [2,4,6]
plt.plot(x, y, color='red', linewidth = 3,marker='o', markerfacecolor='white', markersize=10)
plt.ylim(0,10)
plt.xlim(0,10)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Display marker')
plt.show()
```

```
#Write a Python program to plot several lines with different format styles in one command using arrays.
import numpy as np
a = np.arange(0, 10, 1)
plt.plot(a, a, 'g-', a, a*2, 'b^', a, a**2, 'rs')
plt.show()
```

```
#Write a Python program to display grid and draw line charts of the closing value of Alphabet Inc.
import datetime as DT
from matplotlib import pyplot as plt
from matplotlib.dates import date2num
data = [(DT.datetime.strptime('03-10-2016', "%d-%m-%Y"), 772.559998),
        (DT.datetime.strptime('04-10-2016', "%d-%m-%Y"), 776.429993),
        (DT.datetime.strptime('05-10-2016', "%d-%m-%Y"), 776.469971),
        (DT.datetime.strptime('06-10-2016', "%d-%m-%Y"), 776.859985),
        (DT.datetime.strptime('07-10-2016', "%d-%m-%Y"), 775.080017 )]
x = [date2num(date) for (date, value) in data]
y = [value for (date, value) in data]
fig = plt.figure()
graph = fig.add_subplot(111)
graph.plot(x,y,'b-o')
graph.set_xticks(x)
graph.set_xticklabels([date.strftime("%d-%m-%Y") for (date, value) in data])
plt.xlabel('Date')
plt.ylabel('Closing Value')
plt.title('Closing stock value of Alphabet Inc.') 
plt.grid(linestyle='-', linewidth='0.5', color='blue')
plt.show()
```
